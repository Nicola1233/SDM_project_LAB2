```python
from rdflib import Graph, Namespace, RDF, RDFS
from rdflib.namespace import XSD


g = Graph()
pub = Namespace("http://www.example.org/publication#")
g.bind("pub", pub)

classes = ["Author","Event","Conference","Workshop","Editor",
           "ConferenceChair","JournalEditor","Reviewer",
           "Journal","Volume","Proceedings","Paper","Keyword"]

for c in classes:
    g.add((pub[c], RDF.type, RDFS.Class))

g.add((pub.Conference,      RDFS.subClassOf, pub.Event))
g.add((pub.Workshop,        RDFS.subClassOf, pub.Event))
g.add((pub.Editor,          RDFS.subClassOf, pub.Author))
g.add((pub.ConferenceChair, RDFS.subClassOf, pub.Editor))
g.add((pub.JournalEditor,   RDFS.subClassOf, pub.Editor))
g.add((pub.Reviewer,        RDFS.subClassOf, pub.Author))

def oprop(uri, dom, ran):
    g.add((uri, RDF.type,  RDF.Property))
    g.add((uri, RDFS.domain, dom))
    g.add((uri, RDFS.range,  ran))

def dprop(uri, dom, dtype):
    g.add((uri, RDF.type, RDF.Property))
    g.add((uri, RDFS.domain, dom))
    g.add((uri, RDFS.range,  dtype))


oprop(pub.cites,            pub.Paper,       pub.Paper)
oprop(pub.writes,           pub.Author,      pub.Paper)
oprop(pub.reviews,          pub.Reviewer,    pub.Paper)
oprop(pub.isCorrespondingAuthor, pub.Author, pub.Paper)

oprop(pub.presentedIn,      pub.Paper,       pub.Proceedings)
oprop(pub.publishedIn,      pub.Paper,       pub.Volume)
oprop(pub.isPartOf,         pub.Proceedings, pub.Event)
oprop(pub.belongsTo,        pub.Volume,      pub.Journal)
oprop(pub.assignedBy,       pub.Reviewer,      pub.Editor)
oprop(pub.edits,            pub.JournalEditor, pub.Journal)
oprop(pub.runs,             pub.ConferenceChair, pub.Event)
oprop(pub.isAbout,          pub.Paper,         pub.Keyword)

dprop(pub.title,        pub.Paper,   XSD.string)
dprop(pub.abstractText, pub.Paper,   XSD.string)
dprop(pub.pages,        pub.Paper,   XSD.integer)
dprop(pub.doi,          pub.Paper,   XSD.string)
dprop(pub.url,          pub.Paper,   XSD.anyURI)
dprop(pub.citationCount,pub.Paper,   XSD.integer)
dprop(pub.name,         pub.Author,  XSD.string)
dprop(pub.email,        pub.Author,  XSD.string)
dprop(pub.hIndex,       pub.Author,  XSD.integer)
dprop(pub.cityLiteral,  pub.Proceedings, XSD.string)
dprop(pub.yearLiteral,  pub.Proceedings, XSD.gYear)
dprop(pub.volumeYear,   pub.Volume,     XSD.gYear)

g.serialize("publication_TBOX.ttl", format="turtle")
print("T-Box written to publication_TBOX.ttl")
```
